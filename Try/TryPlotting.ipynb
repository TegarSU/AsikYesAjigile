```python
import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
```

```python
# #Path

# path_to_predict = "C:/Users/USER/Downloads/Tugas Akhir/AsikYesAjigile/predict/"

path_to_predict = "F:/178/Tugas Akhir/dataset/predict/"
```

```python
def angle(p1, p2):
    ang1 = np.arctan2(*p1[::-1])
    ang2 = np.arctan2(*p2[::-1])
    return np.rad2deg((ang1 - ang2) % (2 * np.pi))

def emotion(a,v):
    if ((v >= 0) & (a >= 0)):
        cluster = 'cluster 1'
    if ((v < 0) & (a >= 0)):
        cluster = 'cluster 2'
    if ((v < 0) & (a <= 0)):
        cluster = 'cluster 3'
    if ((v >= 0) & (a < 0)):
        cluster = 'cluster 4'
        
    return cluster

def get_val(data):
    v_pred = data.loc[:,2] #X
    a_pred = data.loc[:,0] #Y
    
    v_true = data.loc[:,3] #X
    a_true = data.loc[:,1] #Y
    
    return a_pred.tolist(), v_pred.tolist(), a_true.tolist(), v_true.tolist()

def plot_model(pred,true):
    plt.figure(figsize=(18, 6))

    plt.plot(pred, color='b') #Pred
    plt.plot(true, color='r') #True
    plt.show()
    
def plot_scatter(pred_a,pred_v,true_a,true_v):
    plt.figure(figsize=(18, 6))

    plt.scatter(pred_v,pred_a,color='b')
    plt.scatter(true_v,true_a,color='r')
    plt.axhline(y=0,color='black', label ="X")
    plt.axvline(x=0,color='black', label ="Y")
    plt.show()

def evaluate(true,pred):
    acc = accuracy_score(true, pred)
    rec = recall_score(true, pred, average='micro')
    prec = precision_score(true, pred, average='micro')
    mat = confusion_matrix(true, pred)
    print("\nAccuracy\t:",acc.round(2) * 100)
    print("\nRecall\t:",rec.round(2) * 100)
    plt.matshow(mat)
    plt.colorbar()
    plt.show()
```

# CASE 1

```python
case = 'case1'
data = pd.read_excel(path_to_predict+case+'_norm_svr.xlsx',header=None)
a_pred,v_pred,a_true,v_true = get_val(data)
```

```python
# #Pred
pred = list()
true = list()

for a,v,at,vt in zip(a_pred,v_pred,a_true,v_true):
    pred.append(emotion(a,v))
    true.append(emotion(at,vt))
```

Plotting Valence

```python
plot_model(v_pred,v_true)
```

Plotting Arousal

```python
plot_model(a_pred,a_true)
```

Plotting Valence-Arousal

```python
plot_scatter(a_pred,v_pred,a_true,v_true)
```

Score

```python
evaluate(true,pred)
```
